Retrieve data showing us the best valuable player from calculating the difference between the player's wage and their market value.

```python
df1 = pd.DataFrame(data_frame, columns=['Name', 'Wage', 'Value'])
# converting the str fields to int and removing the 'K', 'M', 'B' symbols
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in  x:
        if len(x) > 1:
            return float(x.replace('B', '')) * 1000000000
        return 0.0
# replacing the € symbol in the wage & value columns   
wage = df1['Wage'].replace('[\€]', '', regex=True).apply(value_to_float)
value = df1['Value'].replace('[\€]', '', regex=True).apply(value_to_float)

df1['Wage'] = wage
df1['Value'] = value

df1['difference'] = df1['Value'] - df1['Wage']

#sort difference from highest to lowest
df1.sort_values('difference', ascending=False)
```

## Data Visualization

Generate a scatter plot to visualize our data above.

Plotting the above in an interactive plot graph using Bokeh

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(Wage,Value)", "(@Wage, @Value)"),
    ("Name", "@Name")]
)

#create a new plot with a title and axis labels
p = figure(title="Soccer 2019", x_axis_label='Wage', y_axis_label='Value', plot_width=700, plot_height=700, tools=[TOOLTIPS])
p.circle('Wage', 'Value', size=10, source=df1)
show(p)
```
